```python
import pandas as pd

data = {
    "review": [
        "ChatGPT is extremely helpful for my studies",
        "Sometimes it gives incorrect answers",
        "Amazing AI tool, very easy to use",
        "Not bad but responses can be slow",
        "Worst experience, totally useless",
        "Very useful for coding and explanations",
        "Needs improvement in accuracy",
        "Excellent for beginners",
        "Average experience, nothing special",
        "I love using ChatGPT every day"
    ],
    "rating": [5, 2, 5, 3, 1, 5, 3, 4, 3, 5]
}

df = pd.DataFrame(data)
df
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

df['rating'].value_counts().plot(kind='bar')
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")
plt.title("Rating Distribution of ChatGPT Reviews")
plt.show()
```

```python
df['rating'].mean()
```

```python
from textblob import TextBlob

df['sentiment_score'] = df['review'].apply(
    lambda x: TextBlob(x).sentiment.polarity
)

df[['review', 'sentiment_score']]
```

```python
df['sentiment'] = df['sentiment_score'].apply(
    lambda x: "Positive" if x > 0 else "Negative" if x < 0 else "Neutral"
)

df[['review', 'sentiment']]
```

```python
df['sentiment'].value_counts().plot(kind='bar')
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.title("Sentiment Analysis of ChatGPT Reviews")
plt.show()
```
